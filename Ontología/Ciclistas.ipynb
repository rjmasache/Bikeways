```python
import sys
from SPARQLWrapper import SPARQLWrapper, JSON
from pandas import json_normalize
import numpy as np
```

```python
endPoint = "http://192.168.1.2:7200/repositories/Bikeways"
sparql = SPARQLWrapper(endPoint)
```

```python
sparql.endpoint
```

```python
    # Consulta SPARQL
    queryStringCiclistas = """
    PREFIX bikeways: <https://bikeways.org/>
    PREFIX schema: <http://schema.org/>
    SELECT ?id ?nombre ?genero ?email ?telefono ?imagen ?peso ?altura ?nacionalidad ?edad
    WHERE {
        ?cyclist a bikeways:Cyclist .
        ?cyclist schema:identifier ?id .
        ?cyclist schema:name ?nombre .
        OPTIONAL { ?cyclist schema:gender ?genero . }
        OPTIONAL { ?cyclist schema:email ?email . }
        OPTIONAL { ?cyclist schema:telephone ?telefono . }
        OPTIONAL { ?cyclist schema:image ?imagen . }
        OPTIONAL { ?cyclist schema:weight ?peso . }
        OPTIONAL { ?cyclist schema:height ?altura .}
        OPTIONAL { ?cyclist schema:nationality ?nacionalidad . }
        OPTIONAL { ?cyclist bikeways:cyclistAge ?edad . }
    } ORDER BY ?name
    """
```

```python
sparql.setQuery(queryStringCiclistas)
sparql.setReturnFormat(JSON)

results = (sparql.query().convert())
```

```python
results = json_normalize(results['results']['bindings'])
```

```python
results
```

```python
# Eliminar columnas innecesarias
results = results.drop(columns=['id.type', 'nombre.type', 'genero.type', 'email.type', 'telefono.datatype', 'telefono.type', 'imagen.type', 'peso.type', 'altura.datatype', 'altura.type', 'nacionalidad.type', 'edad.datatype', 'edad.type'])
```
